```python
from sklearn.model_selection import train_test_split
features_without_date_and_targetlabel = features.drop(['Date','Z_Score_HelpfulVotes'],axis=1)
training_data,test_data = train_test_split(features_without_date_and_targetlabel,test_size=0.2,random_state=42)
training_labels,test_labels = train_test_split(features['Z_Score_HelpfulVotes'],test_size=0.2,random_state=42)
#feature labels will be the target for models
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()

lin_reg.fit(training_data,training_labels)
```

```python
# test_predictions=lin_reg.predict(test_data)
# test_predictions
# Computed below
```

The below code is to predict the Z_Score_HelpfulVotes values of some of the instances from test data and display both
predicted values and the actual values.

Using Linear Regression Model

```python
some_data = test_data.iloc[:25]
some_labels = test_labels.iloc[:25]
some_predictions = lin_reg.predict(some_data)

# print('predictions:',some_predictions)
# print(some_labels)
# print(some_predictions)

test_df = pd.DataFrame()
test_df['Labels']=some_labels
test_df['Predictions'] = some_predictions
test_df
```

#The above are the predictions from Linear Regression

```python
some_labels = pd.DataFrame(some_labels)
```



Now, calculating the Root Mean Square Error for the predictions that have been made on the test data



```python
from sklearn.metrics import mean_squared_error
test_predictions = lin_reg.predict(test_data)
lin_mse = mean_squared_error(test_labels,test_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

Decision Tree Regression

Comparitively powerfull than Linear Regression

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(training_data, training_labels)
```

```python
tree_test_predictions = tree_reg.predict(test_data)
tree_mse = mean_squared_error(test_labels,tree_test_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

```python
some_tree_predictions = tree_reg.predict(some_data)

# print('predictions:',some_predictions)
# print(some_labels)
# print(some_predictions)

tree_test_df = pd.DataFrame()
tree_test_df['Labels']=some_labels
tree_test_df['Predictions'] = some_tree_predictions
test_df
```
